```python
from dateutil.parser import parse

from dateutil.parser import parse

def fix_timeformat(df):
    # Function to add seconds to time if missing
    def add_seconds_if_missing(time_str):
        if len(time_str) == 5:  # Format HH:MM
            return time_str + ':00'
        return time_str
    df.rename(columns={'Time': 'time'}, inplace=True)
    # Function to parse dates in different formats
    def parse_date(date_str):
        return parse(date_str, dayfirst=True)

    # Fix '24:00' and '24:00:00' time entries
    df.loc[df['time'] == '24:00', 'time'] = '00:00:00'
    df.loc[df['time'] == '24:00:00', 'time'] = '00:00:00'
    
    # Apply the function to add seconds if missing
    df['time'] = df['time'].apply(add_seconds_if_missing)
    # Parse the 'Date' column to handle different formats
    df['Date'] = df['Date'].apply(parse_date)
    
    # Combine 'Date' and 'time' columns into a new 'datetime' column
    df['Date and time'] = pd.to_datetime(df['Date'].dt.strftime('%Y-%m-%d') + ' ' + df['time'], format='%Y-%m-%d %H:%M:%S')
    
    
    return df
```

```python

def add_date_details(df):
    # Ensure the 'datetime' column exists and is of datetime type
    if 'Date and time' not in df.columns:
        raise ValueError("DataFrame must contain a 'Date and time' column")
    if not pd.api.types.is_datetime64_any_dtype(df['Date and time']):
        raise ValueError("'Date and time' column must be of Date and time type")
    
    # Create new columns for day, weekday, month, and season
    df['weekday'] = df['Date and time'].dt.weekday + 1  # Monday is 0 and Sunday is 6; we add 1 to match isoweekday
    df['month'] = df['Date and time'].dt.month
    df['hour'] = df['Date and time'].dt.hour
    
    # Determine the season based on the month
    def get_season(month):
        if month in [12, 1, 2]:
            return 1  # Winter
        elif month in [3, 4, 5]:
            return 2  # Spring
        elif month in [6, 7, 8]:
            return 3  # Summer
        elif month in [9, 10, 11]:
            return 4  # Autumn
    
    df['season'] = df['month'].apply(get_season)
    
    return df
```

```python
def plot_time_plots(df, plot_columns, save_path='plots/',plot_name= 'time_plots.png'):
    import os
    columns = plot_columns
    # Calculate averages for each column
    hourly_avg = df.groupby('hour')[columns].mean()
    weekday_avg = df.groupby('weekday')[columns].mean()
    monthly_avg = df.groupby('month')[columns].mean()
    seasonal_avg = df.groupby('season')[columns].mean()

    num_columns = len(columns)
    fig, axs = plt.subplots(num_columns, 4, figsize=(20, 5 * num_columns))

    for i, column in enumerate(columns):
        # Plot hourly average
        axs[i, 0].bar(hourly_avg.index, hourly_avg[column], color='skyblue')
        axs[i, 0].set_title(f'Hourly Average {column}')
        axs[i, 0].set_xlabel('Hour')
        axs[i, 0].set_ylabel(f'Average {column}')

        # Plot weekday average
        axs[i, 1].bar(weekday_avg.index, weekday_avg[column], color='lightgreen')
        axs[i, 1].set_title(f'Weekday Average {column}')
        axs[i, 1].set_xlabel('Weekday')
        axs[i, 1].set_ylabel(f'Average {column}')

        # Plot monthly average
        axs[i, 2].bar(monthly_avg.index, monthly_avg[column], color='lightcoral')
        axs[i, 2].set_title(f'Monthly Average {column}')
        axs[i, 2].set_xlabel('Month')
        axs[i, 2].set_ylabel(f'Average {column}')

        # Plot seasonal average
        axs[i, 3].bar(seasonal_avg.index, seasonal_avg[column], color='orange')
        axs[i, 3].set_title(f'Seasonal Average {column}')
        axs[i, 3].set_xlabel('Season')
        axs[i, 3].set_ylabel(f'Average {column}')

    plt.tight_layout()
    
    # Ensure the save directory exists
    os.makedirs(save_path, exist_ok=True)
    
    # Save the figure
    plt.savefig(os.path.join(save_path, plot_name))
    plt.show()
```

```python
def fill_nan_with_interpolation(df, max_gap=12):
    def interpolate_column(col):
        # Identify the indices where NaN values exist
        nan_indices = np.where(col.isna())[0]

        # Process each segment of consecutive NaNs
        i = 0
        while i < len(nan_indices):
            start_idx = nan_indices[i]
            end_idx = start_idx

            # Find the end of the consecutive NaNs
            while end_idx + 1 in nan_indices:
                end_idx += 1

            # Number of consecutive NaNs
            gap_length = end_idx - start_idx + 1

            if gap_length <= max_gap:
                # Calculate the difference between the values before and after the NaNs
                lower_idx = start_idx - 1
                upper_idx = end_idx + 1

                if lower_idx >= 0 and upper_idx < len(col):
                    lower_value = col.iloc[lower_idx]
                    upper_value = col.iloc[upper_idx]

                    if not pd.isna(lower_value) and not pd.isna(upper_value):
                        # Fill the gap with linear interpolation
                        step = (upper_value - lower_value) / (gap_length + 1)
                        for j in range(gap_length):
                            interpolated_value = lower_value + (j + 1) * step
                            interpolated_value = max(interpolated_value, col.min())
                            interpolated_value = min(interpolated_value, col.max())
                            col.iloc[start_idx + j] = interpolated_value

            # Move to the next segment
            i = end_idx + 1

        return col

    # Apply the function to each column in the DataFrame
    return df.apply(interpolate_column)
```

```python

def replace_outliers_with_nearest(df, lower_quantile=0.0005, upper_quantile=0.995):
    df_cleaned = df.copy()
    for column in df.columns:
        if np.issubdtype(df[column].dtype, np.number):  # Only process numeric columns
            lower_threshold = df[column].quantile(lower_quantile)
            upper_threshold = df[column].quantile(upper_quantile)

            for i in range(len(df)):
                if df.iloc[i][column] < lower_threshold:
                    nearest_value = df[df[column] >= lower_threshold][column].min()
                    df_cleaned.loc[df_cleaned.index[i], column] = nearest_value
                elif df.iloc[i][column] > upper_threshold:
                    nearest_value = df[df[column] <= upper_threshold][column].max()
                    df_cleaned.loc[df_cleaned.index[i], column] = nearest_value
    return df_cleaned
```

```python
from sklearn.linear_model import Ridge
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer, SimpleImputer
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import FunctionTransformer
from sklearn.compose import TransformedTargetRegressor
```

```python
# Function to clip values to the range [min_value, max_value]
def clip_values(X, min_value, max_value):
    return np.clip(X, min_value, max_value)

def impute_missing_values(df, features):
    df_imputed = df.copy()
    
    for column in df.columns:
        if column in features and np.issubdtype(df[column].dtype, np.number):  # Only process numeric columns
            known = df[df[column].notnull()]
            unknown = df[df[column].isnull()]
            
            if not unknown.empty:
                # Remove the target column from the features list
                training_features = [f for f in features if f != column]
                
                # Create a pipeline with SimpleImputer and Ridge regression
                model = make_pipeline(
                    SimpleImputer(strategy='mean'),
                    Ridge(alpha=1.0)
                )
                
                # Wrap the model in TransformedTargetRegressor to apply the clip_values function
                min_value = df[column].min()
                max_value = df[column].max()
                
                model = TransformedTargetRegressor(
                    regressor=model,
                    func=lambda x: clip_values(x, min_value, max_value),
                    inverse_func=lambda x: clip_values(x, min_value, max_value)
                )
                
                # Filter out rows where the target column is not null
                known_features = df[df[column].notnull()][training_features]
                known_targets = df[df[column].notnull()][column]
                
                # Use the original features for the unknown rows
                unknown_features = df[df[column].isnull()][training_features]
                
                if known_features.shape[0] > 1 and unknown_features.shape[0] > 0:
                    model.fit(known_features, known_targets)
                    
                    predictions = model.predict(unknown_features)
                    
                    # Align predictions with the original DataFrame indices
                    df_imputed.loc[unknown_features.index, column] = predictions
                else:
                    print(f"Not enough data to impute {column}.")
            else:
                print(f"No missing values in {column}.")
    
    # Apply IterativeImputer as a final step
    imputer = IterativeImputer(random_state=0, min_value=0)
    df_final = pd.DataFrame(imputer.fit_transform(df_imputed), columns=df_imputed.columns, index=df_imputed.index)
    
    return df_final
```

```python
def wind_direction_to_cardinal(df, column_name):
    #Converts wind direction in degrees to cardinal direction.
    #'N' =1, 'NE' = 2, 'E'= 3, 'SE'= 4, 'S' =5, 'SW'= 6, 'W' = 7, 'NW' =8
    def degrees_to_cardinal(degrees):
        #Converts degrees to one of the 8 major cardinal directions.
        
        
        if degrees is None:
            return None
        directions = [1, 2, 3, 4, 5, 6, 7, 8, 1]
        idx = int((degrees + 22.5) // 45)
        return directions[idx]
    
    df['Wind Cardinal Direction'] = df[column_name].apply(degrees_to_cardinal)
    return df
```
